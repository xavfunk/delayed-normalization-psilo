```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
## inputs
durations = [0, 17, 33, 67, 134, 267, 533]
fixed_duration = 134
total_duration = 1200 # in ms 
var_duration = np.vstack([np.hstack((np.ones(duration), np.zeros(total_duration - duration))) for duration in durations])
var_isi = np.vstack([np.hstack((np.ones(fixed_duration), np.zeros(duration), np.ones(fixed_duration), np.zeros(total_duration- duration - 2*fixed_duration))) for duration in durations])
```

- make a nice grid for including the noise patterns
- animate numerator and denominator seperately for CTS!

```python
fig, axs = plt.subplots(7, 2, figsize = (12,6))
# xdata, ydata, zdata = [], [], []
# ln, = ax.plot([], [], 'ro')

# for ax in axs[:,0]:
#     vline = ax.axvline(x=0, color='r', linestyle='--')
vlines = {}

for i, ax in enumerate(axs[:,0]):
    vlines[i] = ax.axvline(x=0, color='r', linestyle='--')
    ax.plot(var_isi[i], color = 'black')
    
    
def init(inputs=var_isi):
    for i, ax in enumerate(axs[:, 0]):

        ax.set_xlim(0, 1000)
        ax.set_ylim(-0.1, 1.1)
#         vlines[i] = ax.axvline(x=0, color='r', linestyle='--')

    return

def update(frame):
#     xdata.append(frame)
#     ydata.append(np.sin(frame))
#     zdata.append(frame)
#     ln.set_data(xdata, ydata)
    for i, ax in enumerate(axs[:,0]):
        vlines[i].set_xdata(frame)
    return 

ani = FuncAnimation(fig, update, frames=np.linspace(0, 1000, 64),
                    init_func=init, blit=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, axs = plt.subplots(2)
xdata, ydata, zdata = [], [], []

ln, = axs[0].plot([], [], 'ro')
vline = axs[0].axvline(x=0, color='r', linestyle='--')

ln_1, = axs[1].plot([], [], 'ro')
vline_1 = axs[1].axvline(x=0, color='r', linestyle='--')


def init():
    for ax in axs.flatten():
    
        ax.set_xlim(0, 2*np.pi)
        ax.set_ylim(-1, 1)
        return ln,

def update(frame):
    for ax in axs.flatten():
        xdata.append(frame)
        ydata.append(np.sin(frame))
    #     zdata.append(frame)
        ln.set_data(xdata, ydata)
        vline.set_xdata(frame)
        return ln, 

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)
plt.show()
```
