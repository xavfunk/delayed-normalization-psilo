```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Only draw when needed

```python
data_only = pd.read_csv('logs/sub-02_frametimings_onlyDrawWhenNeeded.csv')
```

```python
data_only.shape
```

```python
all_diffs_only = np.zeros((data_only.shape[0]-1, data_only.shape[1]))
for i, col in enumerate(data_only.columns):
    all_diffs_only[:, i] = data_only[col].diff()[1:] 
    plt.hist(data_only[col].diff(), alpha = .5)
```

```python
plt.hist(all_diffs_only.flatten())
```
